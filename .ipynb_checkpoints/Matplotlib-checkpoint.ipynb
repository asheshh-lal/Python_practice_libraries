# Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x=np.linspace(0,5,11)
```

```python
y=x**2
```

```python
x
```

```python
y
```

```python
#Functional method
plt.plot(x,y)
plt.xlabel("xlabel")
plt.ylabel("ylabel")
plt.title("TESTABLE")
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
##Object oriented method
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel("x-label")
axes.set_ylabel("y-label")
axes.set_title("second_test")
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
axes1.plot(x,y)
axes1.set_title("large")
axes2.plot(y,x)
axes2.set_title("small")
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

#axes.plot(x,y)

for current_ax in axes:
        current_ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

#axes.plot(x,y)

axes[0].plot(x,y)
axes[0].set_xlabel("small")
axes[1].plot(y,x)
axes[1].set_ylabel("small2")
plt.tight_layout()
```

# figure size and DPI


```python
fig=plt.figure(figsize=(8,2))
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes=plt.subplots(nrows=3,ncols=1,figsize=(8,2),dpi=200)
axes[0].plot(x,y)
axes[1].plot(y,x)
axes[2].plot(x,y)
plt.tight_layout()
```

```python
#save a figure
fig.savefig('my_picture.png',dpi=200)
```

```python
fig = plt.figure()
ax= fig.add_axes([0,0,1,1])
ax.plot(x,y**2,label='x square')
ax.plot(y,x**3,label='x cube')
ax.legend(loc=10)
```

# Plot appearence

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='green',linewidth=3,alpha = 0.1,linestyle='-.')
```
