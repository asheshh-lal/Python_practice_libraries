```python
unique_players = df['player_name'].unique()

player_label_dict = {}

for players in unique_players:
    labels_for_players = df.loc[df['player_name']==players, 'label'].unique()
    player_label_dict[players] = labels_for_players[0]
    
player_df = pd.DataFrame(list(player_label_dict.items()), columns=['player_name', 'label'])
player_df
```
