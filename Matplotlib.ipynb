# Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x=np.linspace(0,5,11)
```

```python
y=x**2
```

```python
x
```

```python
y
```

```python
#Functional method
plt.plot(x,y)
plt.xlabel("xlabel")
plt.ylabel("ylabel")
plt.title("TESTABLE")
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
##Object oriented method
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel("x-label")
axes.set_ylabel("y-label")
axes.set_title("second_test")
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
axes1.plot(x,y)
axes1.set_title("large")
axes2.plot(y,x)
axes2.set_title("small")
```
