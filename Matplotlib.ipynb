# Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline #using matplotlib you dont need to use plt.show()
```

```python
import numpy as np
```

```python
x=np.linspace(0,5,11)
```

```python
y=x**2
```

```python
x
```

```python
y
```

```python
#Functional method
plt.plot(x,y)
plt.xlabel("xlabel")
plt.ylabel("ylabel")
plt.title("TESTABLE")
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
##Object oriented method
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel("x-label")
axes.set_ylabel("y-label")
axes.set_title("second_test")
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
axes1.plot(x,y)
axes1.set_title("large")
axes2.plot(y,x)
axes2.set_title("small")
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

#axes.plot(x,y)

for current_ax in axes:
        current_ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

#axes.plot(x,y)

axes[0].plot(x,y)
axes[0].set_xlabel("small")
axes[1].plot(y,x)
axes[1].set_ylabel("small2")
plt.tight_layout()
```

# figure size and DPI


```python
fig=plt.figure(figsize=(8,2))
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes=plt.subplots(nrows=3,ncols=1,figsize=(8,2),dpi=200)
axes[0].plot(x,y)
axes[1].plot(y,x)
axes[2].plot(x,y)
plt.tight_layout()
```

```python
#save a figure
fig.savefig('my_picture.png',dpi=200)
```

```python
fig = plt.figure()
ax= fig.add_axes([0,0,1,1])
ax.plot(x,y**2,label='x square')
ax.plot(y,x**3,label='x cube')
ax.legend(loc=10)
```

# Plot appearence

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='green',linewidth=3,alpha = 0.5,linestyle='--',marker="o",ms=10,
        markerfacecolor='yellow',markeredgewidth=3,markeredgecolor='blue')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='green',linewidth=3,alpha = 0.5,linestyle='--')
ax.set_xlim([0,1])
ax.set_ylim([0,2])
```

```python
import numpy as np
x = np.arange(0,100)
y=x*2
z=x**2
```

```python
x
```

```python
y
```

```python
z
```

```python
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_xlabel("x-label")
ax.set_ylabel("y-label")
ax.set_title("exercise")
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([0.2,0.5,.2,.2])
axes1.plot(x,y,color='red')
axes2.plot(x,y,color='red')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([0.2,0.5,.4,.4])
#large
axes1.plot(x,z)
axes1.set_xlabel("X-label")
axes1.set_ylabel("Y-label")

#small
axes2.plot(x,y)
axes2.set_xlabel("x-label")
axes2.set_ylabel("y-label")
axes2.set_xlim([20,22])
axes2.set_ylim([30,50])
```

```python
fig,axes = plt.subplots(1,2,figsize=(12,2))
axes[0].plot(x,y, color ='blue',lw=3,ls='--')
axes[1].plot(x,z, color ='red',lw=3)
```
