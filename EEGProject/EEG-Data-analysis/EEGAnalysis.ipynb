```python
df = data.loc[:,['Sensor Time','Lft X Pos', 'Lft Y Pos', 'Lft Pupil Diameter', 'Rt X Pos', 'Rt Y Pos',]]
```

```python
df
```

```python
df['gaze_x'] = (df['Lft X Pos'] + df['Rt X Pos'])/2
df['gaze_y'] = (df['Lft Y Pos'] + df['Rt Y Pos'])/2
df['time_convert']=df['Sensor Time']/1000000
```

```python
df
```

```python
def remove_missing_data(df,missing = 0):
    mx=np.array(np.array(df.gaze_x)==missing, dtype =int)
    my=np.array(np.array(df.gaze_y)==missing, dtype =int)
    eye_x = df.gaze_x[mx != 1]
    eye_y = df.gaze_x[my != 1]
    time = df['time_convert'][mx != 1]
    return eye_x,eye_y,time
```

```python
eye_x , eye_y, time = remove_missing_data(df)
```

```python
time.count()
```

```python
result = pd.concat([eye_x, eye_y, time], axis=1)
```

```python
result
```

```python
def fixation_detection(df, mindur=15):
    x, y, time = remove_missing_data(df, missing=0)
    Sfix = []
    Efix = []
    fixationstart = False
    for i in range(1, len(x)):
        try:
            gap = abs(int((time[i - 1] - time[i]) * 1000))
            if gap <= mindur and not fixationstart:
                fixationstart = True
                Sfix.append(time[i])
            if gap > mindur and fixationstart:
                fixationstart = False
                Efix.append([Sfix[-1], time[i - 1], int(time[i - 1] - Sfix[-1]), x[i], y[i]])
        except KeyError:
            pass
    return Efix
```

```python
fixation_detection(df,15)
```
